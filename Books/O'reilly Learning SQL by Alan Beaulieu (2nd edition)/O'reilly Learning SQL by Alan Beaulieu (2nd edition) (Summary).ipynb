### 2.3.1 - Character Data

Character data can be stored as either fixed-length or variable-length strings/

__Fixed-Lenght__ -> `char(20)` (max lenght is 255 bytes)

__Variable-Lenght__ -> `varchar(20)` (max lenght is 65,535 bytes)

If you need to store longer strings (such as emails, XML, documents, etc.), then you will want to use one of the text types (mediumtext and
